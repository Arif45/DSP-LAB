```python
#moving Average
import os
import sys
import seaborn
from scipy.signal import medfilt


def movingaverage(values,window):
    weigths = np.repeat(1.0, window)/window
    m = np.convolve(values, weigths, 'valid')
    return m


signal = [1,6,3,13,5,6,7,10,9,4,6,5,7]
print(movingaverage(signal,3))
plt.subplot(2,2,1)       
plt.title("Moving Average")
plt.stem(movingaverage(signal,3), 'b')
plt.show()

def median_filter(data):
    return medfilt(data, kernel_size=3)


print(median_filter(signal))
plt.subplot(2,2,1)       
plt.title("Median Filter")
plt.stem(median_filter(signal), 'b')
plt.show()
```

```python
import cmath
import matplotlib.pyplot as plt
def dft(x):
    t = []
    N = len(x)
    for k in range(N):
        a = 0
        for n in range(N):
            a += x[n]*cmath.exp(-2j*cmath.pi*k*n*(1/N))
        t.append(a)
    return t
def idft(t):
    x = []
    N = len(t)
    for n in range(N):
        a = 0
        for k in range(N):
            a += t[k]*cmath.exp(2j*cmath.pi*k*n*(1/N))
        a /= N
        x.append(a)
    return x


plt.title("Main Signal")
plt.stem(y1)
plt.show()
dft=dft(y1)
# print(dft)
plt.title("DFT")
plt.plot(dft)
plt.show()

idft=idft(y1)
# print(dft)
plt.title("IDFT")
plt.plot(idft)
plt.show()
```

```python
import numpy.fft as DFT ### Getting DFT using fft
N = 100
f=1/10
t = np.linspace(0,20,N)
x = np.sin(2*np.pi*f*t) + np.exp(-t/3)*np.cos(2*np.pi*f*t) ##Composite signal

y = DFT.fftn(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = DFT.ifftn(y*window, norm='ortho')

plt.plot(t, x, '-b')
plt.plot(t, y, '-r')
plt.plot(t, yr, 'g')
plt.show()
```

```python
from pylab import *
import scipy.signal as signal

#Plot frequency and phase response
def mfreqz(b,a=1):
    w,h = signal.freqz(b,a)
    h_dB = 20 * log10 (abs(h))
    subplot(211)
    plot(w/max(w),h_dB)
    ylim(-150, 5)
    ylabel('Magnitude (db)')
    xlabel(r'Normalized Frequency (x$\pi$rad/sample)')
    title(r'Frequency response')


n = 61
a = signal.firwin(n, cutoff = 0.3, window = "hamming")
#Frequency and phase response
mfreqz(a)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def shade(imag, percent):
    tinted_imag = imag * (1 - percent)
    return tinted_imag

def vertical_gradient_line(image, reverse=False):
    number_of_columns = image.shape[1]
    if reverse:
        C = np.linspace(1, 0, number_of_columns)
    else:
        C = np.linspace(0, 1, number_of_columns)
    C = np.dstack((C, C, C))
    return C

def horizontal_gradient_line(image, reverse=False):
    number_of_rows, number_of_columns = image.shape[:2]
    C = np.linspace(1, 0, number_of_rows)
    C = C[np.newaxis,:]
    C = np.concatenate((C, C, C)).transpose()
    C = C[:, np.newaxis]
    return C

windmills = plt.imread('s.jpg')
tinted_windmills = shade(windmills, 0.7)
plt.imshow(tinted_windmills)
plt.show()

horizontal_brush = horizontal_gradient_line(windmills)
tinted_windmills =  windmills * windmills
plt.imshow(tinted_windmills)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import pywt
import pywt.data


# Load image
original = pywt.data.camera()

# Wavelet transform of image, and plot approximation and details
titles = ['Approximation', ' Horizontal detail',
          'Vertical detail', 'Diagonal detail']
coeffs2 = pywt.dwt2(original, 'bior1.3')
LL, (LH, HL, HH) = coeffs2
fig = plt.figure(figsize=(12, 3))
for i, a in enumerate([LL, LH, HL, HH]):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
    ax.set_title(titles[i], fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

fig.tight_layout()
plt.show()
```
